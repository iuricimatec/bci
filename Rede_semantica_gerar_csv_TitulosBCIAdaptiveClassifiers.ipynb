```python
# gloassary
# https://images.webofknowledge.com/images/help/WOS/hs_wos_fieldtags.html
def parse_txt2csv(txt_file, csv_file):
    with open(txt_file, 'r', encoding='utf-8') as file:
        content = file.read()

    entries = re.split(r'\n(?=PT J)', content)

    # Create a CSV file and write headers
    with open(csv_file, 'w', newline='', encoding='utf-8') as csv_file:
        csv_writer = csv.writer(csv_file)
        headers = ["PT", "AU", "TI", "SO", "LA", "DT", "DE", "AB", "C1", "C3", "RP", "EM", "FU", "FX", "CR", "NR", "TC", "Z9", "U1", "U2", "PU", "PI", "PA", "SN", "EI", "J9", "JI", "PD", "PY", "VL", "IS", "PG", "WC", "WE", "SC", "GA", "UT", "DA"]
        csv_writer.writerow(headers)

        for entry in entries:
            # Extract values using regular expressions
            pt = re.search(r'PT (.+)', entry)
            au = re.search(r'AU (.+?)\n', entry)
            ti = re.search(r'TI (.+?)\n', entry)
            so = re.search(r'SO (.+?)\n', entry)
            la = re.search(r'LA (.+?)\n', entry)
            dt = re.search(r'DT (.+?)\n', entry)
            de = re.search(r'DE (.+?)\n', entry)
            ab = re.search(r'AB (.+?)\n', entry)
            c1 = re.search(r'C1 (.+?)\n', entry)
            c3 = re.search(r'C3 (.+?)\n', entry)
            rp = re.search(r'RP (.+?)\n', entry)
            em = re.search(r'EM (.+?)\n', entry)
            fu = re.search(r'FU (.+?)\n', entry)
            fx = re.search(r'FX (.+?)\n', entry)
            cr = re.search(r'CR (.+?)\n', entry)
            nr = re.search(r'NR (.+?)\n', entry)
            tc = re.search(r'TC (.+?)\n', entry)
            z9 = re.search(r'Z9 (.+?)\n', entry)
            u1 = re.search(r'U1 (.+?)\n', entry)
            u2 = re.search(r'U2 (.+?)\n', entry)
            pu = re.search(r'PU (.+?)\n', entry)
            pi = re.search(r'PI (.+?)\n', entry)
            pa = re.search(r'PA (.+?)\n', entry)
            sn = re.search(r'SN (.+?)\n', entry)
            ei = re.search(r'EI (.+?)\n', entry)
            j9 = re.search(r'J9 (.+?)\n', entry)
            ji = re.search(r'JI (.+?)\n', entry)
            pd = re.search(r'PD (.+?)\n', entry)
            py = re.search(r'PY (.+?)\n', entry)
            vl = re.search(r'VL (.+?)\n', entry)
            is_ = re.search(r'IS (.+?)\n', entry)
            pg = re.search(r'PG (.+?)\n', entry)
            wc = re.search(r'WC (.+?)\n', entry)
            we = re.search(r'WE (.+?)\n', entry)
            sc = re.search(r'SC (.+?)\n', entry)
            ga = re.search(r'GA (.+?)\n', entry)
            ut = re.search(r'UT (.+?)\n', entry)
            da = re.search(r'DA (.+?)\n', entry)

            # Write values to CSV row
            row = [pt.group(1) if pt else "",
                   au.group(1) if au else "",
                   ti.group(1) if ti else "",
                   so.group(1) if so else "",
                   la.group(1) if la else "",
                   dt.group(1) if dt else "",
                   de.group(1) if de else "",
                   ab.group(1) if ab else "",
                   c1.group(1) if c1 else "",
                   c3.group(1) if c3 else "",
                   rp.group(1) if rp else "",
                   em.group(1) if em else "",
                   fu.group(1) if fu else "",
                   fx.group(1) if fx else "",
                   cr.group(1) if cr else "",
                   nr.group(1) if nr else "",
                   tc.group(1) if tc else "",
                   z9.group(1) if z9 else "",
                   u1.group(1) if u1 else "",
                   u2.group(1) if u2 else "",
                   pu.group(1) if pu else "",
                   pi.group(1) if pi else "",
                   pa.group(1) if pa else "",
                   sn.group(1) if sn else "",
                   ei.group(1) if ei else "",
                   j9.group(1) if j9 else "",
                   ji.group(1) if ji else "",
                   pd.group(1) if pd else "",
                   py.group(1) if py else "",
                   vl.group(1) if vl else "",
                   is_.group(1) if is_ else "",
                   pg.group(1) if pg else "",
                   wc.group(1) if wc else "",
                   we.group(1) if we else "",
                   sc.group(1) if sc else "",
                   ga.group(1) if ga else "",
                   ut.group(1) if ut else "",
                   da.group(1) if da else ""]

            # Write the row to the CSV file
            csv_writer.writerow(row)
```

### Data pre-process for Document Title.
The method for constructing the proposed semantic networks basically consists of (1) elimination of words without
intrinsic meaning and (2) changing the remaining words to their canonical form, as suggested in Refs. [14,15]. Each title is
a network where all vertices (i.e. words) are interconnected, generating cliques (a clique is a subset of vertices in a graph G
that are mutually adjacent to one another [18]). Words that appear in more than one title are vertices of connection between
the titles. In this way, we construct a semantic network based on papers’ titles published in a given journal.
https://www.sciencedirect.com/science/article/pii/S0378437110010125

- remove contractions
- remove pontuation
- lower case
- remove stop words
- Lemmatization


- R1 Each title consists of one sentence.
- R2 Graphic signs, such as period, semicolon, question mark, exclamation point and ellipses are eliminated.
- R3 Names should form a single word. For instance, ‘‘Bose–Einstein’’ should be converted to ‘‘boseeinstein’’, or ‘‘Albert Einstein’’
should be converted to ‘‘alberteinstein’’.
- R4 Ordinal numbers should be written as follows: ‘‘first’’, ‘‘second’’, etc.
- R5 Numbers should be written textually. For instance, ‘‘onezero’’ in place of ‘‘10’’.
- R6 Composite words should be considered as only one word. For instance, ‘‘Rio de Janeiro’’ should be converted to ‘‘riodejaneiro’’,
or ‘‘e-mail’’ should be converted to ‘‘email’’.
- R7 Words incorrectly spelt, should be corrected.
- R8 Specialized language should be kept as much as possible.
- R9 Words repeated in the same title should be excluded, leaving only one occurrence of the word.
- R10 Word strings that are jointly meaningful, are made into a single word (e.g. blackhole, computerscience.).
- R11 Titles in another language should be translated into the language of analysis (e.g. an article published in a journal whose main
language is Portuguese, with a title in another language should be translated into Portuguese).
- R12 Verbs must be in infinitive form
- R13 Connectives must be removed
- R14 Plural removed

####Remove Stopwords

```python
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in stop_words])
df_semantic['TI'] = df_semantic['TI'].apply(lambda x: remove_stopwords(x))
```

####Lemmatization

```python

nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
def lemmatize_words(text):
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()])
df_semantic['TI'] = df_semantic['TI'].apply(lambda text: lemmatize_words(text))
```

```python
# generate Gephi's input file within network configuration.
def gen_clique_line(data):
  title = data[0]
  y = data[1]
  split_t = title.split(' ')
  Temp_df = pd.DataFrame(columns=['Source','Target','Type','Weight',"Date"])

  #making word combinations and adding as connections
  for comb in itertools.combinations(split_t, 2):
    list_row = [comb[0], comb[1], "undirected",1, y]
    Temp_df.loc[len(Temp_df)] = list_row
  return Temp_df
    # Clique_df.loc[len(Clique_df)] = list_row
```

```python
df_semantic = pd.read_csv(path + '/pre_processed_titles_data20231217.csv')
Clique_df = pd.DataFrame(columns=['TI','X'])

with Pool() as pool:
    # execute tasks in order
    for result in tqdm(pool.imap_unordered(gen_clique_line, zip(df_semantic['TI'],df_semantic["PY"])), total=len(df_semantic)):

        # print(f'Got result: {result}', flush=True)
        Clique_df =  pd.concat([Clique_df,result])



Clique_df.to_csv(path + '/clique_net_titles_20231216.csv',index=False)
```
